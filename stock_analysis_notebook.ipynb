### 1. Problem Definition

**Analytical Problem:** This project analyzes stock market performance of major technology companies (Apple, Microsoft, Google, Amazon) from 2016 to 2026, to help retail investors understand long-term investment returns and market trends.

### 2. Import Libraries

I will use these libraries for data analysis and visualization.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime

# Set plot style
plt.style.use('seaborn-v0_8-whitegrid')

print("Libraries imported successfully!")
```

### 3. Data Loading

**Data Source:** Yahoo Finance (via yfinance library)  
**Date Accessed:** April 16, 2026  
**Time Period:** January 1, 2016 to April 16, 2026  

I chose Yahoo Finance because it is free, reliable, and provides historical stock data.

**Stocks Analyzed:**
- Apple (AAPL)
- Microsoft (MSFT)
- Google (GOOGL)
- Amazon (AMZN)

```python
# Define the stocks we want to analyze
stocks = ['AAPL', 'MSFT', 'GOOGL', 'AMZN']
stock_names = ['Apple', 'Microsoft', 'Google', 'Amazon']

# Set date range
start_date = '2016-01-01'
end_date = '2026-04-16'

# Download stock data from Yahoo Finance
print("Downloading stock data from Yahoo Finance...")
print("This may take a moment...")

data = yf.download(stocks, start=start_date, end=end_date)

print("Data downloaded successfully!")
print(f"Data shape: {data.shape}")
```

```python
# Look at the first few rows
print("First 5 rows of closing prices:")
data['Close'].head()
```

```python
# Check for missing values
print("Checking for missing values:")
print(data['Close'].isnull().sum())
```

### 4. Data Cleaning

The data looks good. There are no missing values, so we don't need to clean it much.

```python
# Extract closing prices
close_prices = data['Close'].copy()

# Simple check - if any NaN values, fill with previous day's price
close_prices = close_prices.fillna(method='ffill')

print("Data cleaning complete!")
print(f"Date range: {close_prices.index[0].strftime('%Y-%m-%d')} to {close_prices.index[-1].strftime('%Y-%m-%d')}")
```

### 5. Basic Statistics

Let's calculate some basic statistics for each stock.

```python
# Calculate basic statistics
stats = pd.DataFrame()

for stock in stocks:
    stats[stock] = [
        close_prices[stock].mean(),
        close_prices[stock].std(),
        close_prices[stock].min(),
        close_prices[stock].max(),
        close_prices[stock].iloc[-1] - close_prices[stock].iloc[0]  # Total change
    ]

stats.index = ['Mean Price', 'Std Dev', 'Min Price', 'Max Price', 'Total Change ($)']
stats.columns = stock_names

print("Basic Statistics:")
round(stats, 2)
```

### 6. Calculate Returns

Now let's calculate the investment returns for each stock.

```python
# Calculate percentage returns
returns = pd.DataFrame()

for i, stock in enumerate(stocks):
    start_price = close_prices[stock].iloc[0]
    end_price = close_prices[stock].iloc[-1]
    pct_return = ((end_price - start_price) / start_price) * 100
    
    returns[stock_names[i]] = [start_price, end_price, pct_return]

returns.index = ['Start Price (2016)', 'End Price (2026)', 'Total Return (%)']

print("Investment Returns (2016-2026):")
round(returns, 2)
```

### 7. Visualization - Stock Prices Over Time

Let's plot the stock prices to see the trends visually.

```python
# Plot stock prices
plt.figure(figsize=(14, 8))

for i, stock in enumerate(stocks):
    plt.plot(close_prices.index, close_prices[stock], label=stock_names[i], linewidth=2)

plt.title('Stock Prices: 2016-2026', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)

# Save the figure
plt.savefig('stock_prices.png', dpi=150, bbox_inches='tight')
plt.show()

print("Figure saved as 'stock_prices.png'")
```

### 8. Calculate Moving Averages

Moving averages help smooth out price fluctuations and show trends more clearly.

```python
# Calculate 50-day and 200-day moving averages for Apple
aapl_prices = close_prices['AAPL'].copy()

# Calculate moving averages
aapl_prices['MA50'] = aapl_prices.rolling(window=50).mean()
aapl_prices['MA200'] = aapl_prices.rolling(window=200).mean()

print("Moving averages calculated!")
```

```python
# Plot Apple with moving averages
plt.figure(figsize=(14, 7))

plt.plot(aapl_prices.index, aapl_prices['AAPL'], label='Apple Price', alpha=0.7)
plt.plot(aapl_prices.index, aapl_prices['MA50'], label='50-Day MA', color='orange')
plt.plot(aapl_prices.index, aapl_prices['MA200'], label='200-Day MA', color='red')

plt.title('Apple Stock Price with Moving Averages', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)

plt.savefig('apple_moving_averages.png', dpi=150, bbox_inches='tight')
plt.show()

print("Figure saved as 'apple_moving_averages.png'")
```

### 9. Compare Stock Performance

Let's compare the performance of all stocks by normalizing their prices.

```python
# Normalize prices to start at 100
normalized = pd.DataFrame()

for stock in stocks:
    normalized[stock] = (close_prices[stock] / close_prices[stock].iloc[0]) * 100

# Plot normalized performance
plt.figure(figsize=(14, 8))

for i, stock in enumerate(stocks):
    plt.plot(normalized.index, normalized[stock], label=stock_names[i], linewidth=2)

plt.title('Stock Performance Comparison (Starting at $100)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Value ($)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)

plt.savefig('performance_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print("Figure saved as 'performance_comparison.png'")
```

### 10. Calculate Volatility

Volatility measures how much a stock price fluctuates. Higher volatility means higher risk.

```python
# Calculate daily returns
daily_returns = close_prices.pct_change()

# Calculate annualized volatility
volatility = daily_returns.std() * np.sqrt(252) * 100  # 252 trading days per year

# Create bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(stock_names, volatility, color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4'])

plt.title('Stock Volatility (Annualized)', fontsize=16)
plt.xlabel('Stock', fontsize=12)
plt.ylabel('Volatility (%)', fontsize=12)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.1f}%', ha='center', va='bottom', fontsize=11)

plt.savefig('volatility_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print("Figure saved as 'volatility_comparison.png'")
```

### 11. Summary Statistics Table

Let's create a final summary table.

```python
# Create summary table
summary = pd.DataFrame()

for i, stock in enumerate(stocks):
    start_price = close_prices[stock].iloc[0]
    end_price = close_prices[stock].iloc[-1]
    total_return = ((end_price - start_price) / start_price) * 100
    ann_return = total_return / 10  # 10 years
    vol = volatility[stock]
    
    summary[stock_names[i]] = [
        f"${start_price:.2f}",
        f"${end_price:.2f}",
        f"{total_return:.1f}%",
        f"{ann_return:.1f}%",
        f"{vol:.1f}%"
    ]

summary.index = ['Start Price (2016)', 'End Price (2026)', 'Total Return', 
                'Avg Annual Return', 'Volatility']

print("=" * 80)
print("STOCK PERFORMANCE SUMMARY (2016-2026)")
print("=" * 80)
print(summary.to_string())
print("=" * 80)
```
